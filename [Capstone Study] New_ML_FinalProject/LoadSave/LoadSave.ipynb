```
import os
import numpy as np
from Preprocessing.Preprocessor import preprocessing
from FeatureExtraction.ExtractFeature import *
import cv2
import h5py
```

```
 class_name = ['B','C','D','E','CT']
```

```

def Load_image(namefile):
    dirPath = '/content/Capstone_Dataset' + '/' + namefile
    images = []
   
    labels = []
    for id,class_ in enumerate(class_name):
        fname = os.path.join(dirPath,class_)
        for filename in os.listdir(fname):
            if filename.endswith('.jpg') or filename.endswith('.JPG'): 
                imagePath = os.path.join(fname,filename)
                img = cv2.imread(imagePath)
                img = preprocessing(img,128,128)
                images.append(img)
                labels.append(id)
    return (np.array(images), np.array(labels))
```

```
data_train, label_train = Load_image('Train')
```

```
print(data_train.shape)
```

```
data_test, label_test = Load_image('Test')
```
